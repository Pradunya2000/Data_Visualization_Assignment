```python
# 1.Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42) 
x = np.random.rand(100) * 100
y = 2.5 * x + np.random.normal(0, 25, 100)

plt.scatter(x, y, color='b', alpha=0.6)

plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.title("Scatter Plot of Synthetic Dataset")

plt.show()
```

```python
# 2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

data = np.random.randn(1000)

plt.hist(data, bins=30, edgecolor='white')

# Add labels and title
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Distribution of Random Numbers")

plt.show()
```

```python
# 3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.

categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [20, 35, 30, 25, 40]

plt.bar(categories, values, color='skyblue')

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Comparison of Categories Based on Values")

plt.show()
```

```python
# 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

np.random.seed(42)
categories = ['Category A', 'Category B', 'Category C', 'Category D']
data = [np.random.normal(loc=mean, scale=5, size=100) for mean in [20, 35, 30, 25]]

plt.boxplot(data, labels=categories, patch_artist=True)

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Distribution of Numerical Values Across Categories")

plt.show()
```

```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)

feature_1 = np.random.rand(100)
feature_2 = 2 * feature_1 + np.random.normal(0, 0.1, 100) 
feature_3 = -1.5 * feature_1 + np.random.normal(0, 0.2, 100)  

data = np.column_stack((feature_1, feature_2, feature_3))

correlation_matrix = np.corrcoef(data, rowvar=False)

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', xticklabels=['Feature 1', 'Feature 2', 'Feature 3'], yticklabels=['Feature 1', 'Feature 2', 'Feature 3'])

plt.title("Correlation Matrix Heatmap")

plt.show()
```
